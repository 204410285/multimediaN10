```python
def factorial2(g):
    if g>0:
        return g*factorial2(g-1)
    else:
        return 1
```

```python
factorial2(5)
```

```python
def fib(n):
    """Print a Fibonacci series up to n."""
    a,b=0,1
    while a<n:
        print(a,end='')
        a,b=b,a+b
        print()
```

```python
fib(2000)
```

```python
from skimage import io,data
import numpy as np

img=io.imread('may.jpg')
rows,cols,dims=img.shape
row2=int(rows/2)
col2=int(cols/2)

img2=img

for i in range(0,row2):
    for j in range(0,cols):
        for k in range(0,3):
            img2[i,j,k]=img[i,j,k]

for i in range(row2,rows):
    for j in range(0,cols):
        for k in range(0,3):
            img2[i,j,k]=img[row2+(-1)*(i-row2),j,k]
            
io.imshow(img2)
io.show()
```
